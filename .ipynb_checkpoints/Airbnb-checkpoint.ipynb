```python
# importing all the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Airbnb_Data.csv")
pd.set_option('display.max_columns', 100) # To view all columns
df.head()
```

```python
print('Rows: ',df.shape[0])
print('Columns: ',df.shape[1])
```

```python
print('Dataframe details: \n')
df.info(verbose=True)
```

```python
df.columns
```

## Basic Visualization of the data for getting valuable insights of the data

### Property Type Distribution

```python
plt.figure(figsize=(12, 10))
property_type_counts = df['property_type'].value_counts()
property_type_counts.plot(kind='bar', color='skyblue')
plt.title('Property Type Distribution')
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

### Price Distribution Histogram

```python
plt.figure(figsize=(10, 6))
plt.hist(df['log_price'], bins=30, color='skyblue', edgecolor='black')
plt.title('Price Distribution Histogram')
plt.xlabel('Log Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Price by room_type

```python
avg_price_by_room_type = df.groupby('room_type')['log_price'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
avg_price_by_room_type.plot(kind='bar', color='lightgreen')
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Log Price')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

### Instant Bookable Property Percentage

```python
plt.figure(figsize=(8, 6))
instant_bookable_counts = df['instant_bookable'].value_counts()
instant_bookable_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightblue'])
plt.title('Instant Bookable Property Percentage')
plt.ylabel('')
plt.show()
```

### Price Trends Over Time as 'first_review' is in datetime format)

```python
df['first_review'] = pd.to_datetime(df['first_review'])
df['first_review'].dt.year
```

```python
df['year'] = df['first_review'].dt.year
avg_price_by_year = df.groupby('year')['log_price'].mean()
plt.figure(figsize=(10, 6))
avg_price_by_year.plot(marker='o', color='orange')
plt.title('Average Price Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Average Log Price')
plt.grid(True)
plt.xticks(avg_price_by_year.index)
plt.show()
```
